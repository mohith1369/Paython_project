# Group By
#### The groupby() function in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty = df.groupby('Size', as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
S_Qty['hue'] = S_Qty['Size']
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Size', y='Qty', data=S_Qty, hue='hue', dodge=False, palette='Set2')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height:.0f}', ha='center', va='bottom')
plt.legend([], [], frameon=False)
```

##### Note: From above Graph you can see that most of the Qty buys M-Size in the sales

```python
category_counts = df['Category'].value_counts()
colors = sns.color_palette('Set2', n_colors=len(category_counts))
plt.figure(figsize=(10, 5))
plt.bar(category_counts.index, category_counts.values, color=colors, edgecolor='black')
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution')
plt.tight_layout()
plt.show()
```

##### Note: From above Graph you can see that most of the  buyers are T-shirt

```python
# Prepare data for scatter plot
sns.scatterplot(data=df, x='Category', y='Size', hue='Category', palette='Set2')
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


df['state_color'] = df['ship-state']

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state', hue='state_color', dodge=False, palette='Set3')
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.legend([], [], frameon=False)  # Hide legend
plt.tight_layout()
plt.show()
df.drop(columns='state_color', inplace=True)
```

```python
top_10_state = df['ship-state'].value_counts().head(10).index

# Filter data
df_top = df[df['ship-state'].isin(top_10_state)].copy()
df_top['color_state'] = df_top['ship-state']  # Use same column for hue

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(data=df_top, x='ship-state', hue='color_state', dodge=False, palette='Set2', order=top_10_state)
plt.title('Top 10 Shipping States')
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend([], [], frameon=False)  # Hide redundant legend
plt.tight_layout()
plt.show()
```

#### Note: From above Graph you can see that most of the  buyers are Maharashtra state
